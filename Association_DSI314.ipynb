```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('/kaggle/input/005414122565/2.csv', names = ['Name'], sep = ',')
df
```

```python
df.shape
```

```python
df.sum
```

```python
data = list(df["Name"].apply(lambda x:x.split(",") ))
data
```

```python
#Let's transform the list, with one-hot encoding
from mlxtend.preprocessing import TransactionEncoder
a = TransactionEncoder()
a_data = a.fit(data).transform(data)
df = pd.DataFrame(a_data,columns=a.columns_)
df = df.replace(False,0)
df
```

```python
first = pd.DataFrame(df.sum() / df.shape[0], columns = ["Support"]).sort_values("Support", ascending = False)
first
```

```python
df = apriori(df, min_support = 0.2 , use_colnames = True, verbose = 1)
```

```python
df_ar = association_rules(df, metric = "confidence", min_threshold = 0.8)
df_ar['union'] = [x | y for x, y in zip(df_ar['antecedents'], df_ar['consequents'])]
df_ar
```

```python
dfs = df_ar.sort_values(by="lift",ascending=False)
```
